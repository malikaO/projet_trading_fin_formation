# Scrapping Data from seekingalpha using Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
import time
from selenium import webdriver
import pandas as pd
```

## Get Data Amazon

```python
browser = webdriver.Chrome(executable_path=r'C:\Users\utilisateur\.wdm\drivers\chromedriver\win32\85.0.4183.87\chromedriver.exe')
company = 'amazon'
browser.get(f'https://seekingalpha.com/search?q={company}&tab=news#10')



#time.sleep(1)
masterdf = pd.DataFrame(columns=["title", "Date", "comment", "texte"])

for page in range(1,10):
    browser.get(" https://seekingalpha.com/search?q=amazon&tab=news#page={}".format(page))
    time.sleep(2)
    page_url = browser.current_url
    temp_title_names = browser.find_elements_by_xpath('//*[@id="result_list"]/ul/li[1]/div[2]/div[1]/a')
    for title in range(len(temp_title_names)):
        tempdf=pd.DataFrame()
        df_titles = browser.find_elements_by_xpath('//*[@id="result_list"]/ul/li[1]/div[2]/div[1]/a')
        element = df_titles[title]
        if element.text not in masterdf.title:
            browser.execute_script("arguments[0].click();", element)
            time.sleep(1.5)
            
            try:  
                Date =browser.find_element_by_xpath('//*[@id="result_list"]/ul/li[2]/div[2]/div[2]').text
                tempdf.loc[0,"Date"] = Date.replace('\n','')
            except NoSuchElementException:
                pass 
            try:  
                comment =browser.find_element_by_xpath('//*[@id="result_list"]/ul/li[1]/div[2]/div[1]/a').text
                tempdf.loc[0,"comment"] = comment.replace('\n','')
            except NoSuchElementException:
                pass 
            try:  
                texte =browser.find_element_by_xpath('//*[@id="result_list"]/ul/li[2]/div[2]/div[3]').text
                tempdf.loc[0,"texte"] = texte.replace('\n','')
            except NoSuchElementException:
                pass 
            
            
            
            
            
            masterdf = masterdf.append(tempdf)
            browser.get(page_url)
            time.sleep(1.5)
        else:
            pass
    
        
masterdf.to_csv("scrape_amazon.csv")
display(masterdf)
```
