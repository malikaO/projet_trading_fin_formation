## Récupération des prix en utilisant Yahoo's Finance API

```python
import yfinance as yf

msft = yf.Ticker("MSFT")

# get stock info
#print(msft.info)
```

## Chargement des données

On travail sur les données des dix dérnières années. 

```python
# Microsoft
tickerSymbol = "MSFT"
tickerData = yf.Ticker(tickerSymbol)
msft = tickerData.history(period='1d', start='2010-10-1', end="2020-10-2")
```

```python
msft
```

```python
# Google
tickerSymbol = "GOOG"
tickerData = yf.Ticker(tickerSymbol)
goog = tickerData.history(period='1d', start='2010-10-1', end="2020-10-2")
```

```python
goog
```

```python
# Amazon
tickerSymbol = "AMZN"
tickerData = yf.Ticker(tickerSymbol)
amzn = tickerData.history(period='1d', start='2010-10-1', end="2020-10-2")
```
